```python
from scipy.ndimage import sobel, laplace, prewitt, gaussian_laplace
from scipy.signal import unit_impulse

def remove_zeros(a: np.ndarray):
    """remove zero-rows and columns, while keeping any if they occur between
    non-zero ones."""
    # Find rows that are not all zero
    non_zero_rows = np.any(a != 0, axis=1)
    # Find columns that are not all zero
    non_zero_cols = np.any(a != 0, axis=0)
    
    # Find the range indices of non-zero rows
    row_indices = np.where(non_zero_rows)[0]
    if len(row_indices) == 0:
        return a[[]]  # There are no non-zero rows
    row_start, row_end = row_indices[0], row_indices[-1] + 1
    
    # Find the range of non-zero columns
    col_indices = np.where(non_zero_cols)[0]
    if len(col_indices) == 0:
        return a[:, []]  # Empty array
    col_start, col_end = col_indices[0], col_indices[-1] + 1
    
    # return all columns and rows in the non-zero range
    return a[row_start:row_end, col_start:col_end]


def sobel_filter():
    ### YOUR CODE HERE
    shape = 7

    # Impule matrix, zeros with a 1 in the center
    center = shape // 2
    impulse = np.zeros((shape, shape))
    impulse[center][center] = 1

    # Apply filter to impulse
    impulse = sobel(impulse)

    # Remove zero columns and rows
    nonzero = remove_zeros(impulse)
    return nonzero

def laplace_filter():
    ### YOUR CODE HERE
    shape = 7

    # Impule matrix, zeros with a 1 in the center
    center = shape // 2
    impulse = np.zeros((shape, shape))
    impulse[center][center] = 1

    # Apply filter to impulse
    impulse = laplace(impulse)

    # Remove zero columns and rows
    nonzero = remove_zeros(impulse)
    return nonzero

def prewitt_filter():
    ### YOUR CODE HERE
    shape = 7

    # Impule matrix, zeros with a 1 in the center
    center = shape // 2
    impulse = np.zeros((shape, shape))
    impulse[center][center] = 1

    # Apply filter to impulse
    impulse = prewitt(impulse)

    # Remove zero columns and rows
    nonzero = remove_zeros(impulse)
    return nonzero


def gaussian_laplace_filter():
    ### YOUR CODE HERE
    shape = 16 # Larger to capture all non-zero values
    sigma=1 # Arbitrary choice


    # Impule matrix, zeros with a 1 in the center
    center = shape // 2
    impulse = np.zeros((shape, shape))
    impulse[center][center] = 1

    # Apply filter to impulse
    impulse = gaussian_laplace(impulse, sigma)

    # Remove zero columns and rows
    nonzero = remove_zeros(impulse)
    return nonzero

# print(sobel_filter())
# print(laplace_filter())
# print(prewitt_filter())
# print(gaussian_laplace_filter())

# Apply filters to Fcam
print(Fcam.shape, sobel_filter().shape)

s_cam = convolve(Fcam, sobel_filter())
l_cam = convolve(Fcam, laplace_filter())
p_cam = convolve(Fcam, prewitt_filter())
g_cam = convolve(Fcam, gaussian_laplace_filter())

plt.subplot(221)
plt.title("Sobel")
plt.imshow(s_cam)

plt.subplot(222)
plt.title("Laplace")
plt.imshow(l_cam)

plt.subplot(223)
plt.title("Prewitt")
plt.imshow(p_cam)

plt.subplot(224)
plt.title("Gaussian")
plt.imshow(g_cam)
```

## Exercise 3: Constructing the Laplace Filter

Utilizing the right and left first-order derivative filters provided earlier, with convolution kernels `{1 -1 0}` and `{0 1 -1}` in both horizontal and vertical directions, can you construct the Laplace filter using these four first-order derivatives? 

*Hint*: Begin by expressing the impulse response as the sum of a horizontal and a vertical function. (In the code cell following this instruction, apply the provided definitions along with the `convolve` function to create the Laplace filter.)


```python
def laplace_derivatives(I):
    """
    Using only the convolve function and these four filters, compute the laplace filter of I.
    The laplace filter is defined as f_xx + f_yy
    """
    Dleft = np.array([[ 0,  1, -1]])
    Dright = np.array([[ 1, -1,  0]])
    Dup = Dleft.T
    Ddown = Dright.T
    
    # approximate f_xx by applying both directions f_x (sequentially)
    f_xx = convolve(convolve(I, Dright), Dleft)

    # approximate f_yy by applying both directions f_y (sequentially)
    f_yy = convolve(convolve(I, Ddown), Dup)

    return f_xx + f_yy

# Test code
# plt.subplot(121)
# plt.imshow(convolve(Fcam, laplace_filter()))
# plt.subplot(122)
# plt.imshow(laplace_derivatives(Fcam))
# print(convolve(Fcam, laplace_filter()) == laplace_derivatives(Fcam))
```

## Exercise 6: Gaussian 2D function

The 2D Gaussian function $G^s(x,y)$ is defined as:
\begin{align}
G^s(x,y) = \frac{1}{s\sqrt{2\pi}} e^{-\frac{x^2+y^2}{2 s^2}}
\end{align}

The function ``Gauss2d(scale, factor)`` should a return a tuple ``(X, Y, G)`` where ``X, Y`` are the 2D grid arrays generated with the ``np.meshgrid`` function and ``G`` is the sampled Gaussian function.

The Gauss kernel should be **normalized** in such a way that at each point in the image it calculates a weighted average of its neighbor points.

```python
def Gauss2d(scale, factor):
    # If we take N = ceil(scale*factor), then a 2N+1 x 2N + 1 filter is a good size for the grid.
    N = np.ceil(scale * factor).astype(int)
    
    ### YOUR CODE HERE

    nx, ny = (2*N +1, 2 * N + 1)
    x = np.linspace(-(nx - 1)/2, (nx - 1)/2, nx)
    y = np.linspace(-(ny - 1)/2, (ny - 1)/2, ny)
    X, Y = np.meshgrid(x, y)
   
    # taken from GeeksforGeeks

    dst = X**2 + Y**2
    
    # Normal part of the Gaussian function
    normal = 1 / (scale * ((2 * np.pi)**0.5))

    # Calculating Gaussian filter
    G = np.exp(-(dst / (2.0 * (scale **2)))) * normal
    
    G = G/np.sum(G)
    return X, Y, G
    
```
